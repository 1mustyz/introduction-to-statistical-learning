```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)

# Bin the 'balance' variable to calculate default rates
bins = np.linspace(0, 2500, 50)  # Create 50 bins from 0 to 2500
df['balance_bin'] = pd.cut(df['balance'], bins=bins, include_lowest=True)

# Calculate the mean default rate for each bin, grouped by student status
default_rates = (df.groupby(['student', 'balance_bin'], observed=True)
                ['default'].mean().reset_index())

# Calculate the midpoints of the bins for plotting
default_rates['balance_mid'] = default_rates['balance_bin'].apply(lambda x: x.mid)

# Plot the default rates for students and non-students
sns.lineplot(data=default_rates[default_rates['student'] == 'Student'],
             x='balance_mid', y='default', color='orange', label='Student')
sns.lineplot(data=default_rates[default_rates['student'] == 'Non-Student'],
             x='balance_mid', y='default', color='blue', label='Non-Student')

# Add horizontal broken lines for the overall default rates
overall_default_rates = df.groupby('student')['default'].mean()
plt.axhline(y=overall_default_rates['Student'], color='orange', linestyle='--', linewidth=1)
plt.axhline(y=overall_default_rates['Non-Student'], color='blue', linestyle='--', linewidth=1)

# Customize the left plot
plt.xlabel('Credit Card Balance')
plt.ylabel('Default Rate')
plt.title('Default Rates by Balance')
plt.legend()
plt.ylim(-0.05, 0.85)  # Adjust y-axis to match the book's style
plt.xlim(0, 2500)

# --- Right Panel: Boxplots of Balance by Student Status ---
plt.subplot(1, 2, 2)

# Create boxplots for balance, separated by student status
sns.boxplot(x='student', y='balance', data=df,
            palette={'Student': 'orange', 'Non-Student': 'blue'})

# Customize the right plot
plt.xlabel('Student Status')
plt.ylabel('Credit Card Balance')
plt.title('Boxplots of Balance by Student Status')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

There is a surprising result here. The p-
values associated with balance and the dummy variable for student status
are very small, indicating that each of these variables is associated with
the probability of default. However, the coefficient for the dummy variable
is negative, indicating that students are less likely to default than non-
students. In contrast, the coefficient for the dummy variable is positive in
Table 4.2. How is it possible for student status to be associated with an
increase in probability of default in Table 4.2 and a decrease in probability
of default in Table 4.3? The left-hand panel of Figure 4.3 provides a graph-
ical illustration of this apparent paradox. The orange and blue solid lines
show the average default rates for students and non-students, respectively,
as a function of credit card balance. The negative coefficient for student in
the multiple logistic regression indicates that for a fixed value of balance
and income, a student is less likely to default than a non-student. Indeed,
we observe from the left-hand panel of Figure 4.3 that the student default
rate is at or below that of the non-student default rate for every value of
balance. But the horizontal broken lines near the base of the plot, which
show the default rates for students and non-students averaged over all val-
ues of balance and income, suggest the opposite effect: the overall student
default rate is higher than the non-student default rate. Consequently, there
is a positive coefficient for student in the single variable logistic regression
output shown in Table 4.2.


In general, the phenomenon seen in Figure 4.3 is known as
confounding.

Multinomial Logistic Regression

multinomial logistic regression, an extension of the binary logistic regression model to handle response variables with more than two classes (K>2K > 2K > 2
). It explains the model formulation, the concept of a baseline class, the interpretation of coefficients, and an alternative coding approach called softmax coding



4.4 Generative Models for Classification

Logistic regression involves directly modeling Pr(Y = k|X = x) using the
logistic function, given by (4.7) for the case of two response classes. In
statistical jargon, we model the conditional distribution of the response Y ,
given the predictor(s) X. We now consider an alternative and less direct
approach to estimating these probabilities. In this new approach, we model
the distribution of the predictors X separately in each of the response
classes (i.e. for each value of Y ). We then use Bayes’ theorem to flip these
around into estimates for Pr(Y = k|X = x). When the distribution of X
within each class is assumed to be normal, it turns out that the model is
very similar in form to logistic regression.
Why do we need another method, when we have logistic regression?
There are several reasons:
• When there is substantial separation between the two classes, the
parameter estimates for the logistic regression model are surprisingly
unstable. The methods that we consider in this section do not suffer
from this problem.
• If the distribution of the predictors X is approximately normal in
each of the classes and the sample size is small, then the approaches
in this section may be more accurate than logistic regression.
• The methods in this section can be naturally extended to the case
of more than two response classes. (In the case of more than two
response classes, we can also use multinomial logistic regression from
Section 4.3.5.)

```python
df['student'] = (df['student'] == 'Student').astype(int)
# Prepare predictors (balance and student)
X = df[['balance', 'student']]
y = df['default']
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Fit LDA
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)

# Predict on training data
y_pred = lda.predict(X)

# Compute confusion matrix
cm = confusion_matrix(y, y_pred)
print("Confusion Matrix:")
print(cm)

# Compute training error rate
error_rate = 1 - accuracy_score(y, y_pred)
print(f"Training Error Rate: {error_rate * 100:.2f}%")

# Detailed metrics
tn, fp, fn, tp = cm.ravel()
print(f"True Negatives (No, No): {tn}")
print(f"False Positives (No, Yes): {fp}")
print(f"False Negatives (Yes, No): {fn}")
print(f"True Positives (Yes, Yes): {tp}")
print(f"Error Rate among Defaulters: {fn / (fn + tp) * 100:.1f}%")
print(f"Error Rate among Non-Defaulters: {fp / (tn + fp) * 100:.1f}%")

```

Improving LDA for Defaulters
To reduce the high FN rate (75.7%) among defaulters, consider:
Adjusting the Decision Threshold: LDA assigns observations to the class with the highest posterior probability (default threshold = 0.5). Since defaulters are rare, lowering the threshold for predicting Yes can increase true positives (TP) at the cost of more false positives (FP).



```python
# Predict probabilities and adjust threshold
y_prob = lda.predict_proba(X)[:, 1]  # Probability of default = Yes
threshold = 0.2  # Lower threshold
y_pred_adjusted = (y_prob > threshold).astype(int)
cm_adjusted = confusion_matrix(y, y_pred_adjusted)
print("Confusion Matrix with Threshold 0.2:")
print(cm_adjusted)
```
